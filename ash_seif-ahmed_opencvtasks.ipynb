```python
#task2
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.mp4', fourcc, 20.0, (640, 480))
key = input("Choose Any of The Following Keys: \nQ- To Terminate The Program\nR- To Rotate The Frame By 90 degree\nZ- To Show Original Frame\nG- To Convert Frame Into GrayScale\nH- To Convert Frame Into HSV\nC- To Save The Frames\nX- to show (grayscale, HSV, rotated and original frames) Together\nS- to save a video stream\n")
i=0
#handle="SsrRzZgGxXhHcC"

while cap.isOpened():
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    rotated = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)

    gray= cv2.cvtColor(gray,cv2.COLOR_GRAY2BGR)
    rotated = img = cv2.resize(rotated, (640,480), interpolation = cv2.INTER_AREA)

    horizontal1 = np.hstack([frame, gray])
    horizontal2 = np.hstack([hsv, rotated])
    vertical= np.vstack([horizontal1,horizontal2])
    vertical= cv2.resize(vertical,(680,600))

    x = cv2.waitKey(1)
    if x != -1:
        key = x

    if ret:
        if key == ord('S') or key == 'S' or key == ord('s') or key == 's':
            out.write(frame)
            cv2.imshow('frame', frame)

        elif key == ord('C') or key == 'C' or key == ord('c') or key == 'c':
            cv2.imwrite('Picture '+str(i)+'.jpg',frame)
            i=i+1
            cv2.imshow('frame', frame)

        elif key == ord('G') or key == 'G' or key == ord('g') or key == 'g':
            cv2.imshow('frame', gray)

        elif key == ord('H') or key == 'H' or key == ord('h') or key == 'h':
            cv2.imshow('frame', hsv)

        elif key == ord('Z') or key == 'Z' or key == ord('z') or key == 'z':
            cv2.imshow('frame', frame)

        elif key == ord('R') or key == 'R' or key == ord('r') or key == 'r':
            cv2.imshow('frame', rotated)

        elif key == ord('X') or key == 'X' or key == ord('x') or key == 'x':
            cv2.imshow('frame', vertical)

        if key == ord('Q') or key == 'Q' or key == ord('q') or key == 'q':
            print("GoodBye!")
            break
       # if (key != ord('x') and key != ord('X') and key != ord('r') and key != ord('R') and key != ord('z') and key != ord('Z') and key != ord('H') and key != ord('h') and key != ord('G') and key != ord('g') and key != ord('c') and key != ord('C') and key != ord('S') and key != ord('s') and key!= -1):
          #  print("GoodBye!")
           # break
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
#task4
import cv2
import numpy as np

circles = []
img = np.zeros((500, 500, 3))


def detecting_mouse(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        circles.append((x, y))
    if event == cv2.EVENT_RBUTTONDOWN:
        if len(circles):
            circles.pop(len(circles) - 1)
            img[:] = 0
    if event == cv2.EVENT_RBUTTONDBLCLK:
        circles.clear()
        img[:] = 0


while True:
    for x in range(0, len(circles)):
        cv2.circle(img, (circles[x][0], circles[x][1]), 25, (0, 0, 255), 1)
    cv2.imshow("image", img)
    cv2.setMouseCallback("image", detecting_mouse)
    if cv2.waitKey(1) == ord('q') or cv2.waitKey(1) == ord('Q'):
        break
cv2.destroyAllWindows()
```

```python
#task5
import cv2
import numpy as np

circles= np.zeros((4,2),np.int64)
counter=0

def mousePoints(events,x,y,flags,params):
    global counter
    if events == cv2.EVENT_LBUTTONDOWN:
        circles[counter]= x,y
        counter = counter+1


img=cv2.imread("jhonsmith.jpg")
while True:
    if counter==4:
        width=250
        height=350

        pts1= np.float32([circles[0],circles[1],circles[2],circles[3]])
        pts2= np.float32([[0,0],[width,0],[0,height],[width,height]])
        matrix=cv2.getPerspectiveTransform(pts1,pts2)
        imgOutput= cv2.warpPerspective(img,matrix,(width,height))
        cv2.imshow("Output Image ",imgOutput)

    for x in range(0,4):
        cv2.circle(img,(circles[x][0],circles[x][1]),3,(0,255,0),cv2.FILLED)

    cv2.imshow("Original Image ",img)
    cv2.setMouseCallback("Original Image ",mousePoints)
    if cv2.waitKey(1) == ord('q') or cv2.waitKey(1) == ord('Q'):
        break
cv2.destroyAllWindows()


```
